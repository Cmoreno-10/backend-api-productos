
## 1. ¿Cuál es el papel de las matrices en el modelo de regresión lineal?

En regresión lineal buscamos encontrar una relación entre variables.

Por ejemplo:

Ingreso = β0 + β1 * Horas de estudio + β2 * Experiencia

Si tenemos muchos datos, escribir todas estas ecuaciones una por una sería muy complejo.

Por eso usamos **matrices**.

En forma matricial el modelo se escribe:

y = Xβ + ε

donde:

- **y** → vector de valores observados
- **X** → matriz de variables (matriz de diseño)
- **β** → vector de coeficientes
- **ε** → error

Las matrices permiten:

- representar **muchas ecuaciones al mismo tiempo**
- resolver el problema de forma **computacional eficiente**
- aplicar **álgebra lineal** para encontrar la mejor solución


```python

# Ejemplo simple

# Matriz de variables (horas de estudio)
X = np.array([
    [1, 2],
    [1, 3],
    [1, 5],
    [1, 7]
])

# Vector de resultados
y = np.array([50, 55, 65, 70])

print("Matriz X:")
print(X)

print("Vector y:")
print(y)
```


## 2. Modelo de mínimos cuadrados en forma matricial

El objetivo de mínimos cuadrados es encontrar los coeficientes β que minimicen el error entre los valores reales y los predichos.

La solución matricial es:

β = (XᵀX)⁻¹ Xᵀy

Aquí aparece la **matriz de diseño X**, que contiene:

- columna de unos (intercepto)
- variables independientes

Esta matriz es fundamental porque representa **toda la información del modelo**.


```python

# Cálculo de mínimos cuadrados con matrices

beta = np.linalg.inv(X.T @ X) @ X.T @ y

print("Coeficientes beta:")
print(beta)
```


## 3. Interpretación geométrica: ortogonalidad y proyección

Geométricamente, mínimos cuadrados significa:

**proyectar el vector y sobre el espacio generado por las columnas de X**.

Esto implica que:

y - Xβ

es **ortogonal** al espacio de X.

Es decir:

Xᵀ(y − Xβ) = 0

Esto garantiza que el error total sea mínimo.


```python

# Verificación de ortogonalidad

beta = np.linalg.inv(X.T @ X) @ X.T @ y

residual = y - X @ beta

print("Vector residual:")
print(residual)

print("Producto X^T * residual (debe ser cercano a 0):")
print(X.T @ residual)
```


## 4. Propiedades de la matriz XᵀX

La matriz XᵀX tiene propiedades importantes:

### Simetría
(XᵀX)ᵀ = XᵀX

### Positividad definida
Para cualquier vector v:

vᵀ(XᵀX)v > 0

Esto es importante porque garantiza que:

- la matriz sea **invertible**
- la solución de mínimos cuadrados sea **única**


```python

A = X.T @ X

print("Matriz X^T X:")
print(A)

print("¿Es simétrica?")
print(np.allclose(A, A.T))
```


## 5. Extensión a mínimos cuadrados ponderados y generalizados

### Mínimos cuadrados ponderados (WLS)

Se utiliza cuando algunos datos tienen **más importancia que otros**.

La solución es:

β = (XᵀWX)⁻¹ XᵀWy

donde W es una matriz de pesos.

### Mínimos cuadrados generalizados (GLS)

Se usa cuando los errores **no tienen varianza constante** o están correlacionados.


```python

# Ejemplo simple de pesos

W = np.diag([1, 1, 2, 2])

beta_w = np.linalg.inv(X.T @ W @ X) @ X.T @ W @ y

print("Coeficientes con pesos:")
print(beta_w)
```


## 6. Simulación de una base de datos

Vamos a simular un escenario real:

Predecir **ingresos mensuales** según:

- años de experiencia
- horas de capacitación


```python

np.random.seed(0)

n = 100

experiencia = np.random.randint(0, 10, n)
capacitacion = np.random.randint(0, 20, n)

ingresos = 1200 + experiencia*400 + capacitacion*150 + np.random.normal(0,300,n)

data = pd.DataFrame({
    "experiencia": experiencia,
    "capacitacion": capacitacion,
    "ingresos": ingresos
})

data.head()
```


## 7. Construcción de la matriz de diseño

La matriz X se construye así:

columna 1 → intercepto  
columna 2 → experiencia  
columna 3 → capacitación


```python

X = np.column_stack((
    np.ones(len(data)),
    data["experiencia"],
    data["capacitacion"]
))

y = data["ingresos"].values

beta = np.linalg.inv(X.T @ X) @ X.T @ y

print("Coeficientes del modelo:")
print(beta)
```


## 8. Visualización de los datos


```python

plt.scatter(data["experiencia"], data["ingresos"])
plt.xlabel("Experiencia")
plt.ylabel("Ingresos")
plt.title("Relación experiencia-ingresos")
plt.show()
```


## 9. Interpretación de los coeficientes

Si el modelo arroja algo cercano a:

β0 ≈ 1200  
β1 ≈ 400  
β2 ≈ 150

Significa:

- cada **año de experiencia** aumenta el ingreso ~400
- cada **hora de capacitación** aumenta el ingreso ~150

Esto permite **tomar decisiones basadas en datos**.



## 10. Validación del modelo

Para evaluar el modelo podemos usar:

- división entrenamiento / prueba
- error cuadrático medio (RMSE)


```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

beta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train

pred = X_test @ beta

rmse = np.sqrt(mean_squared_error(y_test, pred))

print("RMSE del modelo:", rmse)
```
