## **¿Qué es un vector?**

Un vector es una colección de números en un orden definido.\\
Si es una colección de $n$ números, se le llama un vector de $n$ dimensiones.\\

Así, el vector $\vec{A}$ dado por:

$$
\vec{A} =
\begin{bmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{bmatrix}
$$

es un vector columna de $n$ dimensiones con $n$ componentes: $a_1, a_2, \ldots, a_n$.\\

El anterior es un vector columna. Un vector fila $[\vec{B}]$ es de la forma:

$$
\vec{B} = [b_1,\ b_2,\ \ldots,\ b_n]
$$

donde $\vec{B}$ es un vector fila de $n$ dimensiones con componentes $b_1, b_2, \ldots, b_n$.

Supóngase que un punto en el espacio está dado por sus coordenadas $(x, y, z)$.\\
Entonces, si los valores son $x = 3$, $y = 2$, $z = 5$, el vector columna que corresponde a la posición del punto es:

$$
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
3 \\
2 \\
5
\end{bmatrix}
$$

# **Operaciones**
## **Suma**

$$
[A] + [B] =
\begin{bmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{bmatrix}
+
\begin{bmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{bmatrix}
=
\begin{bmatrix}
a_1 + b_1 \\
a_2 + b_2 \\
\vdots \\
a_n + b_n
\end{bmatrix}
$$

Ejemplo

$$
\vec{A} + \vec{B} =
\begin{bmatrix}
2 \\
3 \\
4 \\
1
\end{bmatrix}
+
\begin{bmatrix}
5 \\
-2 \\
3 \\
7
\end{bmatrix}
=
\begin{bmatrix}
2 + 5 \\
3 - 2 \\
4 + 3 \\
1 + 7
\end{bmatrix}
=
\begin{bmatrix}
7 \\
1 \\
7 \\
8
\end{bmatrix}
$$


```python
# Crear vectores desde el origen hasta diferentes puntos
origen = np.array([0, 0])
punto1 = np.array([4, 5])
punto2 = np.array([-4, -5])

# Graficar los vectores
plt.quiver(*origen, *punto1, angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
plt.quiver(*origen, *punto2, angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')

plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.grid()
plt.legend()
plt.title("Vectores desde el Origen")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

## **Resta**

$$
[A] - [B] =
\begin{bmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{bmatrix}
-
\begin{bmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{bmatrix}
=
\begin{bmatrix}
a_1 - b_1 \\
a_2 - b_2 \\
\vdots \\
a_n - b_n
\end{bmatrix}
$$

Ejemplo

$$
\vec{A} - \vec{B} =
\begin{bmatrix}
2 \\
3 \\
4 \\
1
\end{bmatrix}
-
\begin{bmatrix}
5 \\
-2 \\
3 \\
7
\end{bmatrix}
=
\begin{bmatrix}
2 - 5 \\
3 - (-2) \\
4 - 3 \\
1 - 7
\end{bmatrix}
=
\begin{bmatrix}
-3 \\
5 \\
1 \\
-6
\end{bmatrix}
$$

```python
# Crear los vectores a y b
a = np.array([2, 3])
b = np.array([3, 1])
diferencia = a - b

# Crear la figura
fig, ax = plt.subplots(figsize=(7, 7))

# Graficar los vectores a, b y b - a con colores
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='b')
ax.quiver(a[0], a[1], diferencia[0], diferencia[1], angles='xy', scale_units='xy', scale=1, color='red', label='b - a')

# Agregar puntos A y B
ax.scatter(a[0], a[1], color='blue')
ax.text(a[0] + 0.2, a[1], 'A', fontsize=12, color='blue')
ax.scatter(b[0], b[1], color='green')
ax.text(b[0] + 0.2, b[1], 'B', fontsize=12, color='green')

# Agregar líneas punteadas para conectar los puntos
ax.plot([0, a[0]], [0, a[1]], 'b--', linewidth=1)  # Línea al punto A
ax.plot([0, b[0]], [0, b[1]], 'g--', linewidth=1)  # Línea al punto B
ax.plot([a[0], b[0]], [a[1], b[1]], 'r--', linewidth=1)  # Línea entre A y B

# Configurar el gráfico
ax.set_xlim(-1, 6)
ax.set_ylim(-1, 6)
ax.grid()
ax.set_aspect('equal', adjustable='box')
ax.set_title("Resta de Vectores")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.legend()

# Mostrar el gráfico
plt.show()
```

# Multipliación por un escalar

$$
\text{Si } k \text{ es un escalar y } \vec{A} \text{ es un vector de dimensión } n, \text{ entonces}
$$

$$
k\vec{A} = k
\begin{bmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{bmatrix}
=
\begin{bmatrix}
ka_1 \\
ka_2 \\
\vdots \\
ka_n
\end{bmatrix}
$$

Ejemplo

$$
\text{¿Cuál es } 2\vec{A} \text{ si } \vec{A} =
\begin{bmatrix}
25 \\
20 \\
5
\end{bmatrix}
\ ?
$$

$$
2\vec{A} = 2
\begin{bmatrix}
25 \\
20 \\
5
\end{bmatrix}
=
\begin{bmatrix}
2 \times 25 \\
2 \times 20 \\
2 \times 5
\end{bmatrix}
=
\begin{bmatrix}
50 \\
40 \\
10
\end{bmatrix}
$$




```python
# Crear vectores desde el origen hasta diferentes puntos
origen = np.array([0, 0])
punto1 = np.array([4, 5])


# Operación de multiplicación escalar de vectores
escalar = 0.25
multiplicacion_vector1 = escalar * punto1


print("Multiplicación escalar del Vector 1:", multiplicacion_vector1)


# Graficar los vectores
plt.quiver(*origen, *punto1, angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
plt.quiver(*origen, *multiplicacion_vector1, angles='xy', scale_units='xy', scale=1, color='g', label='Vector 1 x Escalar')


plt.xlim(-1, 15)
plt.ylim(-1, 15)
plt.grid()
plt.legend()
plt.title("Multiplicación Escalar de Vectores")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

## **`Combinación lineal`**

$$
\text{Dado } \vec{A}_1, \vec{A}_2, \ldots, \vec{A}_m \text{ como } m \text{ vectores de la misma dimensión } n,
$$

$$
\text{y si } k_1, k_2, \ldots, k_m \text{ son escalares, entonces}
$$

$$
k_1\vec{A}_1 + k_2\vec{A}_2 + \cdots + k_m\vec{A}_m
$$

$$
\text{es una combinación lineal de los } m \text{ vectores}.
$$


Ejemplo: Encontrar las combinaciones lineales

$$
\vec{A} = \begin{bmatrix} 2 \\ 3 \\ 6 \end{bmatrix}, \quad
\vec{B} = \begin{bmatrix} 1 \\ 1 \\ 2 \end{bmatrix}, \quad
\vec{C} = \begin{bmatrix} 10 \\ 1 \\ 2 \end{bmatrix}
$$

Para determinar si \textbf{existe una combinación lineal entre los vectores}
$$\vec{A}, \vec{B}, \vec{C},$$
lo que se busca es saber si alguno de ellos puede escribirse como una combinación lineal de los otros dos. Es decir, por ejemplo:

¿Existe una combinación de la forma

$$
\vec{C} = \alpha \vec{A} + \beta \vec{B}
$$

para algunos escalares reales $\alpha$ y $\beta$?

Lo que nos da el siguiente sistema de ecuaciones:

$$
\left\{
\begin{aligned}
2\alpha + \beta &= 10 \\
3\alpha + \beta &= 1 \\
6\alpha + 2\beta &= 2
\end{aligned}
\right.
$$

Resolver el sistema

Restamos la primera ecuación de la segunda:

$$
(3\alpha + \beta) - (2\alpha + \beta) = 1 - 10 \\
\alpha = -9
$$

Sustituimos $\alpha = -9$ en la primera ecuación:

$$
2(-9) + \beta = 10 \Rightarrow -18 + \beta = 10 \Rightarrow \beta = 28
$$

Verificamos en la tercera ecuación:

$$
6(-9) + 2(28) = -54 + 56 = 2
$$


Sí, existe combinación lineal:

$$
\vec{C} = -9 \vec{A} + 28 \vec{B}
$$


```python
import numpy as np
import plotly.graph_objects as go

# Definir un conjunto de vectores en R^3
v1 = np.array([2, 1, 3])
v2 = np.array([-1, 2, 0])
v3 = np.array([1, -1, 2])

# Definir los coeficientes de la combinación lineal
c1, c2, c3 = 1.5, -0.5, 2  # Modifica estos valores según desees

# Calcular la combinación lineal
linear_combination = c1 * v1 + c2 * v2 + c3 * v3

# Determinar automáticamente el tamaño de los ejes basado en los vectores
max_x = max(abs(v1[0]), abs(v2[0]), abs(v3[0]), abs(linear_combination[0])) + 1
max_y = max(abs(v1[1]), abs(v2[1]), abs(v3[1]), abs(linear_combination[1])) + 1
max_z = max(abs(v1[2]), abs(v2[2]), abs(v3[2]), abs(linear_combination[2])) + 1

# Permitir valores personalizados de los ejes (modifica según sea necesario)
axis_length_x = max_x  # Puedes cambiar estos valores manualmente
axis_length_y = max_y
axis_length_z = max_z

# Crear la figura
fig = go.Figure()

# Agregar los ejes coordenados con los valores personalizados
fig.add_trace(go.Scatter3d(x=[-axis_length_x, axis_length_x], y=[0, 0], z=[0, 0], mode='lines',
                           line=dict(color='black', width=4), name='Eje X'))
fig.add_trace(go.Scatter3d(x=[0, 0], y=[-axis_length_y, axis_length_y], z=[0, 0], mode='lines',
                           line=dict(color='black', width=4), name='Eje Y'))
fig.add_trace(go.Scatter3d(x=[0, 0], y=[0, 0], z=[-axis_length_z, axis_length_z], mode='lines',
                           line=dict(color='black', width=4), name='Eje Z'))

# Agregar los vectores base
fig.add_trace(go.Scatter3d(x=[0, v1[0]], y=[0, v1[1]], z=[0, v1[2]],
                           mode='lines+markers', line=dict(color='blue', width=6),
                           marker=dict(size=6, color='blue'), name='v1'))
fig.add_trace(go.Scatter3d(x=[0, v2[0]], y=[0, v2[1]], z=[0, v2[2]],
                           mode='lines+markers', line=dict(color='green', width=6),
                           marker=dict(size=6, color='green'), name='v2'))
fig.add_trace(go.Scatter3d(x=[0, v3[0]], y=[0, v3[1]], z=[0, v3[2]],
                           mode='lines+markers', line=dict(color='orange', width=6),
                           marker=dict(size=6, color='orange'), name='v3'))

# Agregar el vector resultante (combinación lineal)
fig.add_trace(go.Scatter3d(
    x=[0, linear_combination[0]], y=[0, linear_combination[1]], z=[0, linear_combination[2]],
    mode='lines+markers',
    line=dict(color='red', width=8),
    marker=dict(size=6, color='red'),
    name='Combinación Lineal'
))

# Configurar la vista y mejorar el diseño
fig.update_layout(
    title="Combinación Lineal de Vectores en R³",
    scene=dict(
        xaxis=dict(title='X', range=[-axis_length_x, axis_length_x]),
        yaxis=dict(title='Y', range=[-axis_length_y, axis_length_y]),
        zaxis=dict(title='Z', range=[-axis_length_z, axis_length_z]),
    ),
    margin=dict(l=10, r=10, b=10, t=40),
    showlegend=True
)

# Mostrar la gráfica interactiva
fig.show()
```

## Linealmente independiente

Un conjunto de vectores  
$$
\vec{A}_1, \vec{A}_2, \dots, \vec{A}_m
$$  
se considera linealmente independiente si  

$$
k_1 \vec{A}_1 + k_2 \vec{A}_2 + \cdots + k_m \vec{A}_m = \vec{0}
$$

tiene únicamente la solución  

$$
k_1 = k_2 = \cdots = k_m = 0
$$

**Ejemplo**

¿Son los tres vectores  
$$
\vec{A}_1 = \begin{bmatrix} 25 \\ 64 \\ 144 \end{bmatrix}, \quad
\vec{A}_2 = \begin{bmatrix} 5 \\ 8 \\ 12 \end{bmatrix}, \quad
\vec{A}_3 = \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}
$$  
linealmente independientes?

**Solución**

Escribiendo la combinación lineal de los tres vectores:

$$
k_1 \begin{bmatrix} 25 \\ 64 \\ 144 \end{bmatrix} +
k_2 \begin{bmatrix} 5 \\ 8 \\ 12 \end{bmatrix} +
k_3 \begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix} =
\begin{bmatrix} 0 \\ 0 \\ 0 \end{bmatrix}
$$

Esto da:

$$
\begin{bmatrix}
25k_1 + 5k_2 + k_3 \\
64k_1 + 8k_2 + k_3 \\
144k_1 + 12k_2 + k_3
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$

Las ecuaciones anteriores tienen solo una solución:

$$
k_1 = k_2 = k_3 = 0
$$

Las ecuaciones anteriores son:

$$
25k_1 + 5k_2 + k_3 = 0 \quad \text{(1)}
$$
$$
64k_1 + 8k_2 + k_3 = 0 \quad \text{(2)}
$$
$$
144k_1 + 12k_2 + k_3 = 0 \quad \text{(3)}
$$


Por lo tanto, la única solución es:   $$k_1 = k_2 = k_3 = 0$$

Los tres vectores, por lo tanto, son linealmente independientes.

```python
import numpy as np
import plotly.graph_objects as go

# Definir los vectores
v1 = np.array([25, 64, 144])
v2 = np.array([5, 8, 12])
v3 = np.array([1, 1, 1])

# Crear figura 3D
fig = go.Figure()

# Añadir vectores desde el origen
fig.add_trace(go.Scatter3d(
    x=[0, v1[0]], y=[0, v1[1]], z=[0, v1[2]],
    mode='lines+markers',
    marker=dict(size=4, color='crimson'),
    line=dict(width=6, color='crimson'),
    name='v1 = [25, 64, 144]'
))

fig.add_trace(go.Scatter3d(
    x=[0, v2[0]], y=[0, v2[1]], z=[0, v2[2]],
    mode='lines+markers',
    marker=dict(size=4, color='seagreen'),
    line=dict(width=6, color='seagreen'),
    name='v2 = [5, 8, 12]'
))

fig.add_trace(go.Scatter3d(
    x=[0, v3[0]], y=[0, v3[1]], z=[0, v3[2]],
    mode='lines+markers',
    marker=dict(size=4, color='royalblue'),
    line=dict(width=6, color='royalblue'),
    name='v3 = [1, 1, 1]'
))

# Configuración de la escena
fig.update_layout(
    title='Vectores en 3D (interactivo)',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        xaxis=dict(range=[0, 30]),
        yaxis=dict(range=[0, 70]),
        zaxis=dict(range=[0, 150])
    ),
    legend=dict(x=0.02, y=0.98),
    width=800,
    height=700,
    margin=dict(l=0, r=0, b=0, t=30)
)

# Mostrar figura
fig.show()
```

##  **La norma**

La norma de un vector es una medida de su magnitud o longitud en el espacio. En otras palabras, indica qué tan “largo” es el vector, sin importar su dirección.

\text{Dado un vector}

$$
\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix} \in \mathbb{R}^n,
$$

la norma euclidiana (también llamada norma 2 o norma L2) se define como:

$$
\|\vec{v}\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}
$$

Si v = \begin{bmatrix} 3 \\ 4 \end{bmatrix}, entonces su norma es:

$$
\|\vec{v}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import math

# ---- Definición de vectores ----
# Vector en R^n
a = 3  # Componente en X
b = 4  # Componente en Y
v = np.array([a, b])  # Vector principal

# Otro vector para cálculo del producto punto y ángulo
u = np.array([3, 4])  # Vector u
w = np.array([4, -3])  # Vector w

# ---- Cálculos de la norma (longitud) del vector ----
# Norma del vector usando el teorema de Pitágoras
norma_pitagoras = np.sqrt(np.sum(v**2))

# Norma del vector usando el producto punto consigo mismo
norma_producto_punto = np.sqrt(np.dot(v, v))

# ---- Producto punto y ángulo entre dos vectores ----
# Producto punto entre u y w
producto_punto = np.dot(u, w)

# Normas de los vectores u y w
norma_u = np.linalg.norm(u)
norma_w = np.linalg.norm(w)

# Calcular el coseno del ángulo entre u y w
cos_theta = producto_punto / (norma_u * norma_w)

# Calcular el ángulo en radianes y convertir a grados
theta = math.acos(cos_theta)
theta_grados = np.degrees(theta)

# ---- Visualización gráfica del vector principal ----
fig, ax = plt.subplots(figsize=(6, 6))

# Dibujar los ejes coordenados
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Dibujar el vector principal
ax.quiver(0, 0, a, b, angles='xy', scale_units='xy', scale=1, color='blue', label=f'Vector v ({a}, {b})')

# Dibujar las proyecciones sobre los ejes
ax.plot([0, a], [0, 0], 'r--', label=f'Proyección en X ({a})')  # Línea en X
ax.plot([a, a], [0, b], 'g--', label=f'Proyección en Y ({b})')  # Línea en Y

# Configuración de la gráfica
ax.set_xlim(0, max(a, b) + 2)
ax.set_ylim(0, max(a, b) + 2)
ax.set_aspect('equal')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title(f'Representación geométrica del vector y su norma\n Norma: {norma_pitagoras:.2f}')
ax.legend()

# Mostrar la gráfica
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

# ---- Resultados ----
# Imprimir resultados de la norma del vector principal
print(f"Vector principal v: {v}")
print(f"Norma de v (Teorema de Pitágoras): {norma_pitagoras:.2f}")
print(f"Norma de v (Producto punto): {norma_producto_punto:.2f}\n")

# Imprimir resultados del producto punto y ángulo
print(f"Vector u: {u}")
print(f"Vector w: {w}")
print(f"Producto punto (u · w): {producto_punto}")
print(f"Norma de u: {norma_u:.2f}")
print(f"Norma de w: {norma_w:.2f}")
print(f"Coseno del ángulo (cos θ): {cos_theta:.2f}")
print(f"Ángulo entre u y w: {theta_grados:.2f}°")
```

Un vector unitario $$\vec{U}$$ se define como:

$$
\vec{U} =
\begin{bmatrix}
u_1 \\
u_2 \\
\vdots \\
u_n
\end{bmatrix}
$$

donde

$$
\sqrt{u_1^2 + u_2^2 + u_3^2 + \cdots + u_n^2} = 1
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Definición del vector ----
v = np.array([3, 4])  # Vector inicial (puedes cambiar las componentes)

# ---- Cálculo de la norma del vector ----
norma = np.linalg.norm(v)

# ---- Normalización del vector ----
# Un vector unitario tiene norma igual a 1
vector_unitario = v / norma

# ---- Representación gráfica ----
fig, ax = plt.subplots(figsize=(6, 6))

# Dibujar el círculo unitario
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--', label='Círculo unitario')
ax.add_artist(circle)

# Dibujar los ejes coordenados
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Dibujar el vector original
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'Vector original v ({v[0]}, {v[1]})')

# Dibujar el vector unitario
ax.quiver(0, 0, vector_unitario[0], vector_unitario[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'Vector unitario u ({vector_unitario[0]:.2f}, {vector_unitario[1]:.2f})')

# Configuración de la gráfica
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Normalización de un vector y el círculo unitario')
ax.legend()

# Mostrar la gráfica
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

# ---- Resultados ----
print(f"Vector original: {v}")
print(f"Norma del vector original: {norma:.2f}")
print(f"Vector unitario: {vector_unitario}")
print(f"Norma del vector unitario: {np.linalg.norm(vector_unitario):.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Definición de puntos o vectores ----
a = np.array([2, 3])  # Punto o vector a (x1, y1)
b = np.array([5, 7])  # Punto o vector b (x2, y2)

# ---- Cálculo de la distancia euclidiana ----
# Fórmula: d(a, b) = sqrt((x2 - x1)^2 + (y2 - y1)^2)
distancia = np.linalg.norm(a - b)

# ---- Visualización gráfica en R^2 ----
fig, ax = plt.subplots(figsize=(6, 6))

# Dibujar los puntos
ax.scatter(a[0], a[1], color='blue', label=f'Punto a ({a[0]}, {a[1]})', zorder=5)
ax.scatter(b[0], b[1], color='red', label=f'Punto b ({b[0]}, {b[1]})', zorder=5)

# Dibujar la línea que representa la distancia entre los puntos
ax.plot([a[0], b[0]], [a[1], b[1]], color='green', linestyle='--', label=f'Distancia: {distancia:.2f}')

# Etiquetas de los puntos
ax.text(a[0], a[1], '  a', fontsize=12, color='blue')
ax.text(b[0], b[1], '  b', fontsize=12, color='red')

# Configuración de la gráfica
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_xlim(min(a[0], b[0]) - 1, max(a[0], b[0]) + 1)
ax.set_ylim(min(a[1], b[1]) - 1, max(a[1], b[1]) + 1)
ax.set_aspect('equal')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Distancia Euclidiana entre dos puntos en R²')
ax.legend()

# Mostrar la gráfica
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

# ---- Resultados ----
print(f"Punto a: {a}")
print(f"Punto b: {b}")
print(f"Distancia euclidiana entre a y b: {distancia:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Definición de los vectores ----
u = np.array([2, 1])  # Vector u (base)
v = np.array([3, 4])  # Vector v (proyectado)

# ---- Cálculo de la proyección de v sobre u ----
# Producto punto de u con v
producto_punto = np.dot(u, v)

# Producto punto de u consigo mismo
producto_u_u = np.dot(u, u)

# Verificación: u no debe ser el vector nulo
if producto_u_u == 0:
    raise ValueError("El vector u no puede ser el vector nulo.")

# Proyección de v sobre u
proyeccion = (producto_punto / producto_u_u) * u

# ---- Visualización gráfica ----
fig, ax = plt.subplots(figsize=(6, 6))

# Dibujar los ejes coordenados
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Dibujar los vectores u y v
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f'Vector u ({u[0]}, {u[1]})')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label=f'Vector v ({v[0]}, {v[1]})')

# Dibujar la proyección de v sobre u
ax.quiver(0, 0, proyeccion[0], proyeccion[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'Proyección de v sobre u ({proyeccion[0]:.2f}, {proyeccion[1]:.2f})')

# Dibujar una línea punteada desde v a su proyección
ax.plot([v[0], proyeccion[0]], [v[1], proyeccion[1]], 'r--', label='Línea perpendicular')

# Configuración de la gráfica
ax.set_xlim(-1, max(v[0], proyeccion[0]) + 2)
ax.set_ylim(-1, max(v[1], proyeccion[1]) + 2)
ax.set_aspect('equal')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Proyección de un vector sobre otro en R²')
ax.legend()

# Mostrar la gráfica
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()

# ---- Resultados ----
print(f"Vector u: {u}")
print(f"Vector v: {v}")
print(f"Proyección de v sobre u: {proyeccion}")
```
